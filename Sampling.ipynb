df.groupby --> Pandas function to group a dataframe based on one or more columns. Commonly used in aggregations, summaries, and group-wise operations
Syntax - df.groupby(by) ----> by: the column(s) to group by
Examples - 
Group by "sex" and count the number of passengers ---> grouped = df.groupby("sex")["age"].count()

Group by "sex" and "class", then find mean age ---> grouped = df.groupby(["sex", "class"])["age"].mean()

Find the mean and median of age for each class ---> grouped = df.groupby("class")["age"].agg(["mean", "median"])

The group_keys parameter in df.groupby() controls whether the group labels (keys) should be included in the result as an index when performing group operations.  

When to Use group_keys=False:  
When you want to keep the original DataFrame format.  
When you don’t need hierarchical indexing.  
Useful for flattening the output after using apply().



```python
prop_stratified_sampling = df.groupby("origin", group_keys=False ).apply(lambda x: x.sample(frac=0.25)) # groups the DataFrame (df) by the "origin" column the randomly selects 25% of the rows from each group.
```
