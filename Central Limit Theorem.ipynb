sns.displot() - Distribution Plot in Seaborn  
seaborn.displot() is used to visualize the distribution of a dataset. It can create histograms, kernel density estimates (KDE), and more.

Example -   
Load example dataset  
df = sns.load_dataset("penguins")  

Create a histogram of flipper lengths  
sns.displot(df["flipper_length_mm"])  

kind parameter--->	Plot type: "hist", "kde", "ecdf"  

```python
sns.displot(df["displacement"], kind = "kde")
```

```python
sns.displot(df["displacement"], kind = "hist")
```

```python
sns.displot(df["displacement"], kind = "ecdf")
```

```python
# Now creating a list of sample means where each sample size is 20 and the number of samples considered are 30

sample_mean_list = []

for i in range(0,30):
    random_sample = df["displacement"].sample(20)
    sample_mean_list.append(random_sample.mean())

sample_mean_list
```

```python
print(sample_mean_list)
```

```python
# Plotting the distribution of the sample means

sns.displot(sample_mean_list, kind="kde")
```

```python
import statistics as st
```

```python
st.mean(sample_mean_list)
```
